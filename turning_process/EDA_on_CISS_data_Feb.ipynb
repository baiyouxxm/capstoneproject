```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
qual =r'C:\Users\xaa2cng\Documents\turning\Turning_data\2020_02_26_28_0.csv'
df1 = pd.read_csv(qual, sep=',')
df1.head()
```

```python
plt.plot(df1['num'])
# naming the x axis 
plt.xlabel('Row Index') 
# naming the y axis 
plt.ylabel('num')
# giving a title to my graph 
plt.title('Variable:num') 
# function to show the plot 
plt.show() 
```

```python
meanfeature=df1.groupby('num', as_index=False)['x'].mean()
meanfeature.head()
```

```python
plt.plot(meanfeature['x'])
# naming the x axis 
plt.xlabel('Row Index') 
# naming the y axis 
plt.ylabel('X_mean')
# giving a title to my graph 
plt.title('Variable: X mean change across different products') 
# function to show the plot 
plt.show() 
```

```python
import pandas as pd
import glob

path = r'C:\Users\xaa2cng\Documents\turning\Turning_data\valid_data' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
